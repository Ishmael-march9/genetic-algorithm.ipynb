```python
#load or make data
def f(x, a, noise=0):
  return a*x + (np.random.random(x.shape)-.5)*noise
x = np.arange(0,10,0.1)
a_best = np.random.random()*5
print('a_best',a_best)
y = f(x, a_best, 3)
plt.figure(figsize=(5,5)) ; plt.xlim(x.min(), x.max()) ; plt.ylim(y.min(), y.max())
plt.scatter(x, y) ; plt.grid(True) ; plt.xlabel('x', fontsize=15) ; plt.ylabel('y', fontsize=15) ; plt.title(rf'$a_b$ = {a_best:.2f}', fontsize=15) ;
```
